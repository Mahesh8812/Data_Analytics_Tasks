```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```python
df = pd.read_csv('Iris (1).csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
plt.plot('SepalLengthCm','Species',data=df)
plt.show()
```

```python
df.groupby('Species').size().plot(kind='bar',
                                 figsize = (6,6),
                                 title = 'TOTAL SPECIES',
                                 color='purple')
plt.show()
```

```python
df.groupby('Species').sum()['SepalLengthCm'].plot(kind='bar',
                                                 figsize=(5,5),
                                                 color='green')
plt.show()
```

```python
df.groupby('Species').mean()['SepalLengthCm'].plot(kind='bar',
                                                 figsize=(5,5),
                                                  color='blue')
plt.show()
```

```python
df.groupby('Id').mean()['SepalLengthCm'].plot(kind='bar',
                                                  color='blue')
plt.show()
```

```python
df.groupby('Id').size().plot(kind='bar',
                                 figsize = (6,6),
                                 title = 'TOTAL Ids',
                                 color='red')
plt.show()
```
